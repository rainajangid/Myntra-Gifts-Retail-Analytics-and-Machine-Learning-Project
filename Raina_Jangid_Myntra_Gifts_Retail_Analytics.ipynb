## 3. ***Data Wrangling***

### Data Wrangling Code

````python

# Step 1: Remove Duplicates
df = df.drop_duplicates()

# Step 2: Drop rows with missing CustomerID
df = df[df['CustomerID'].notnull()]

# Step 3: Filter out rows with Quantity or UnitPrice <= 0
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Step 4: Reset index to avoid view-related issues
df = df.reset_index(drop=True)

# Step 5: Convert InvoiceDate to datetime
df.loc[:, 'InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Step 6: Create Revenue column
df.loc[:, 'Revenue'] = df['Quantity'] * df['UnitPrice']

# Step 7: Extract temporal features
df.loc[:, 'Year'] = df['InvoiceDate'].dt.year
df.loc[:, 'Month'] = df['InvoiceDate'].dt.month
df.loc[:, 'Day'] = df['InvoiceDate'].dt.day
df.loc[:, 'Hour'] = df['InvoiceDate'].dt.hour
df.loc[:, 'DayOfWeek'] = df['InvoiceDate'].dt.dayofweek

````

### What all manipulations have you done and insights you found?

Duplicates Removed â†’ ensured no exact row repetitions for accurate stats

Datetime Parsed â†’ enabled trend and seasonality analysis via Year, Month, Hour, etc.

Missing Values Handled â†’ focused only on known CustomerID values (filter out anonymous/guest purchases)

Invalid Transactions Filtered â†’ removed returns/cancellations (Quantity â‰¤ 0) and data entry errors (UnitPrice â‰¤ 0)

Revenue Calculated â†’ added Revenue feature = Quantity * UnitPrice for business performance tracking

Feature Engineering â†’ extracted temporal attributes (like weekday, hour) for demand forecasting or pattern discovery

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothetical Statement - 1
Customers from the United Kingdom spend more per order than customers from other countries.


Hypothetical Statement - 2

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hâ‚€: There is no significant difference in the average revenue per order between UK and non-UK customers.

Hâ‚: UK customers spend more per order than non-UK customers.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind


uk_revenue = df[df['Country'] == 'United Kingdom'].groupby('InvoiceNo')['Revenue'].sum()
non_uk_revenue = df[df['Country'] != 'United Kingdom'].groupby('InvoiceNo')['Revenue'].sum()
stat1, p_value_1 = ttest_ind(uk_revenue, non_uk_revenue, equal_var=False)


print("\nHypothesis 1: UK vs Non-UK Avg Revenue per Order")
print(f"P-Value: {p_value_1:.5f}")
if p_value_1 < 0.05:
    print("Result: Reject the null hypothesis.")
    print(" Interpretation: UK customers spend significantly more per order.")
else:
    print(" Result: Fail to reject the null hypothesis.")
    print(" Interpretation: No significant difference in order value between UK and other countries.")
````

##### Which statistical test have you done to obtain P-Value?

Independent Two-Sample T-Test

##### Why did you choose the specific statistical test?

â†’ We're comparing means between two independent groups: UK vs Non-UK customers on total order value.

P-Value = 0.1705


â†’ Conclusion: Fail to reject the null hypothesis.
No statistically significant difference found in average spending between UK and non-UK customers.



### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hâ‚€: There is no significant difference in quantity purchased between weekdays and weekends.

Hâ‚: Weekdays see significantly higher quantities purchased.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
weekday_qty = df[df['DayOfWeek'] < 5]['Quantity']
weekend_qty = df[df['DayOfWeek'] >= 5]['Quantity']
stat2, p_value_2 = ttest_ind(weekday_qty, weekend_qty, equal_var=False)

print("\nHypothesis 2: Weekday vs Weekend Quantity")
print(f"P-Value: {p_value_2:.5f}")
if p_value_2 < 0.05:
    print(" Result: Reject the null hypothesis.")
    print(" Interpretation: More items are purchased on weekdays than weekends.")
else:
    print(" Result: Fail to reject the null hypothesis.")
    print(" Interpretation: No significant difference in item quantity by day of week.")
````

##### Which statistical test have you done to obtain P-Value?

â†’ Independent Two-Sample T-Test

##### Why did you choose the specific statistical test?

â†’ We're comparing the mean number of items bought across two groups (weekday vs weekend).

P-Value = 9.44e-14
â†’ Conclusion: Reject the null hypothesis.
There is a statistically significant increase in quantity purchased on weekdays vs weekends.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hâ‚€: There is no significant difference in unit price between high-revenue (top 10%) and other orders.

Hâ‚: High-revenue orders have higher unit prices.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value


threshold = df['Revenue'].quantile(0.9)
high_revenue_prices = df[df['Revenue'] >= threshold]['UnitPrice']
low_revenue_prices = df[df['Revenue'] < threshold]['UnitPrice']
stat3, p_value_3 = ttest_ind(high_revenue_prices, low_revenue_prices, equal_var=False)


print("\nHypothesis 3: High Revenue vs Low Revenue Unit Price")
print(f"P-Value: {p_value_3:.5f}")
if p_value_3 < 0.05:
    print(" Result: Reject the null hypothesis.")
    print(" Interpretation: High-revenue orders are associated with significantly higher-priced products.")
else:
    print(" Result: Fail to reject the null hypothesis.")
    print(" Interpretation: No pricing difference detected between high and low revenue orders.")
````

##### Which statistical test have you done to obtain P-Value?

â†’ Independent Two-Sample T-Test

##### Why did you choose the specific statistical test?

â†’ Comparing mean unit price between two groups: top 10% revenue vs rest.

P-Value = 0.00

There is a significant difference â€” high-revenue orders tend to have higher unit prices.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
import os
import tempfile
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingRandomSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.experimental import enable_halving_search_cv  # noqa

#STEP 1: Clean Training & Test Data
def sanitize(X, y):
    mask = np.isfinite(X).all(axis=1) & np.isfinite(y)
    return X[mask], y[mask]

X_train, y_train = sanitize(X_train, y_train)
X_test, y_test = sanitize(X_test, y_test)

# STEP 2: Set safe temp directory for joblib
os.environ['JOBLIB_TEMP_FOLDER'] = tempfile.mkdtemp(dir="/tmp")

# STEP 3: Define safe hyperparameter space
param_dist = {
    'n_estimators': [100, 150, 200],
    'max_depth': [10, 20, 30, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],  # Removed None
    'bootstrap': [True, False]
}

# STEP 4: Create RandomForest model
model = RandomForestRegressor(random_state=42)

# STEP 5: Setup HalvingRandomSearchCV
search = HalvingRandomSearchCV(
    estimator=model,
    param_distributions=param_dist,
    resource='n_samples',
    factor=2,
    min_resources=256,
    max_resources='auto',
    scoring='r2',
    cv=3,
    n_jobs=-1,
    random_state=42,
    verbose=2,
    error_score='raise'
)

# STEP 6: Fit hyperparameter search
search.fit(X_train, y_train)

# STEP 7: Retrieve & refit best model
best_model = search.best_estimator_
best_model.fit(X_train, y_train)

# STEP 8: Evaluate on test set
y_pred = best_model.predict(X_test)
print("Best Parameters:", search.best_params_)
print(" Best R2 Score (CV):", search.best_score_)
print(" Test RÂ²:", r2_score(y_test, y_pred))
print(" Test MSE:", mean_squared_error(y_test, y_pred))
print(" Test MAE:", mean_absolute_error(y_test, y_pred))
````

##### Which hyperparameter optimization technique have you used and why?

We used HalvingRandomSearchCV, a successive halving-based hyperparameter optimization technique. It extends RandomizedSearchCV by allocating increasing resources to better-performing candidates in a staged elimination process. Compared to both GridSearchCV and RandomizedSearchCV, it is significantly more efficient when dealing with large search spaces and high data volumes.
It strikes an optimal trade-off between computational cost and performance by focusing resources only on promising configurations during each iteration.
